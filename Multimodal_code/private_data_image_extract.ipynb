```python
!pip install pytesseract pillow opencv-python numpy torch tensorflow transformers


import os
import cv2
import csv
import pytesseract
from PIL import Image

# Function to preprocess the image
def preprocess_final(im):
    im = cv2.bilateralFilter(im, 5, 55, 60)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    _, im = cv2.threshold(im, 240, 255, 1)
    return im

# Function to extract text from an image
def extract_text(image_path, custom_config=r"--oem 3 --psm 11 -c tessedit_char_whitelist= 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz '"):
    img = cv2.imread(image_path)
    img = preprocess_final(img)
    text = pytesseract.image_to_string(img, lang='eng', config=custom_config)
    return text.replace('\n', ' ')

# Path to the folder containing images
image_folder_path = 'C:/Users/israt/Downloads/private_imagedata_for_code'
extracted_text_csv = 'C:/Users/israt/Downloads/private_imagedata_for_code/extracted_text.csv'

# Extract text from each image and save it in a CSV file
with open(extracted_text_csv, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Image Name', 'Extracted Text'])

    for folder in ['0', '1', '2', '3']:
        subfolder_path = os.path.join(image_folder_path, folder)
        for image_name in os.listdir(subfolder_path):
            image_path = os.path.join(subfolder_path, image_name)
            if os.path.isfile(image_path):
                try:
                    text = extract_text(image_path)
                    writer.writerow([image_name, text])
                    print(f"Processed {image_name}")
                except Exception as e:
                    print(f"Error processing {image_name}: {e}")

print("Text extraction complete. Data saved.")
```

```python
import pandas as pd
import string
data = pd.read_csv(extracted_text_csv)


# Display the first few rows and the summary of null values
data_head = data.head()
null_summary = data.isnull().sum()
data_head, null_summary




# Remove rows with null values in 'Extracted Text'
data = data.dropna(subset=['Extracted Text'])

# Data cleaning: trim whitespace and convert to lowercase
data['Extracted Text'] = data['Extracted Text'].str.strip().str.lower()

# Text processing: Tokenization and removing punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

data['Extracted Text'] = data['Extracted Text'].apply(remove_punctuation)

# Display the first few rows of the cleaned data
print(data.head())

# Save the cleaned data back to CSV
cleaned_file_path = 'C:/Users/israt/Downloads/private_imagedata_for_code/cleaned_data.csv'
data.to_csv(cleaned_file_path, index=False)

print(f"Cleaned data saved to {cleaned_file_path}")
```
